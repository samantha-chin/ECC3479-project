# Pre-Trend Test for DiD Causal Claim

This notebook implements an identification-relevant robustness check for the causal DiD specification. The main DiD estimate is a negative Treat x Post coefficient of approximately -0.075. For this causal interpretation to be credible, treated and control suburbs should not show strong evidence of different pre-treatment trends before the post period begins in 2022.

## Test Design

This notebook runs two robustness checks:

The main models include log population, suburb fixed effects, and year fixed effects.

- Alternative functional forms for the main Treat x Post DiD effect, including log, levels, quadratic, IHS, and interaction specifications.
- A pre-trend test that uses 2017 as the milestone year, estimating treatment-by-year interactions for 2015 and 2016 to evaluate parallel trends before the 2017 milestone.

```python
from pathlib import Path
import sys
import pandas as pd
import statsmodels.formula.api as smf

project_root = Path.cwd()
sys.path.append(str(project_root / "code"))

from robustness_pretrend_test import (
    RESULTS_DIR,
    prepare_data,
    fit_pretrend_model,
    build_pretrend_table,
    joint_pretrend_test,
    build_notes_and_interpretation,
    dataframe_to_markdown,
)
from robustness_functional_form import (
    prepare_data as prepare_ff_data,
    fit_models as fit_ff_models,
    build_table as build_ff_table,
    build_notes_and_interpretation as build_ff_notes_and_interpretation,
)
from robustness_alternative_inference import build_panel

# Original pretrend test using the existing 2021 reference design

df = prepare_data()
result = fit_pretrend_model(df)
pretrend_table = build_pretrend_table(result)
joint_test = joint_pretrend_test(result)
notes, interpretation = build_notes_and_interpretation(joint_test)

# Alternative functional-form robustness checks
ff_df = prepare_ff_data()
ff_results = fit_ff_models(ff_df)
ff_table = build_ff_table(ff_results)
ff_notes, ff_interpretation = build_ff_notes_and_interpretation(ff_results)

# Pretrend test using 2017 as the milestone year
panel_df = build_panel().copy()
pre_years_2017 = [2015, 2016]
for year in pre_years_2017:
    panel_df[f"treated_x_{year}"] = (
        panel_df["treated"] * (panel_df["year"] == year).astype(int)
    )

pretrend_2017_formula = (
    "log_price ~ "
    + " + ".join(f"treated_x_{year}" for year in pre_years_2017)
    + " + log_population + C(suburb) + C(year)"
)
pretrend_2017_result = smf.ols(pretrend_2017_formula, data=panel_df).fit(
    cov_type="cluster", cov_kwds={"groups": panel_df["suburb"]}
)

rows_2017 = []
for year in pre_years_2017:
    coef = pretrend_2017_result.params[f"treated_x_{year}"]
    se = pretrend_2017_result.bse[f"treated_x_{year}"]
    t_stat = pretrend_2017_result.tvalues[f"treated_x_{year}"]
    p_value = pretrend_2017_result.pvalues[f"treated_x_{year}"]
    rows_2017.append(
        {
            "Pre-treatment year": year,
            "Relative to": 2017,
            "Coefficient": f"{coef:.3f}",
            "Std. Error": f"({se:.3f})",
            "t-statistic": f"{t_stat:.3f}",
            "p-value": f"{p_value:.3f}",
        }
    )

rows_2017.append(
    {
        "Pre-treatment year": "N",
        "Relative to": "",
        "Coefficient": int(pretrend_2017_result.nobs),
        "Std. Error": "",
        "t-statistic": "",
        "p-value": "",
    }
)
pretrend_table_2017 = pd.DataFrame(rows_2017)

wald_2017 = pretrend_2017_result.wald_test(
    " = 0, ".join(f"treated_x_{year}" for year in pre_years_2017) + " = 0",
    scalar=True,
)
joint_test_2017 = {
    "statistic": float(wald_2017.statistic),
    "p_value": float(wald_2017.pvalue),
    "df_num": len(pre_years_2017),
}

pretrend_table, ff_table, pretrend_table_2017
```
