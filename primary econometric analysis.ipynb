# Primary Econometric Analysis


**Analysis type**: Causal

**Research question**

Did exposure to early LXRP construction and completions by 2017 change house prices in treated suburbs relative to suburbs not yet exposed?

**Identification strategy and assumptions**

We use a difference-in-differences (DiD) design. The key identifying assumption is parallel trends: in the absence of early LXRP exposure, house prices in early-exposed suburbs would have evolved in parallel with those in not-yet-exposed suburbs. The panel structure (2015-2024) provides multiple pre-treatment years to assess baseline comparability and reduces reliance on any single pre-period. Common time shocks are controlled for via time indicators, and time-varying demographics are controlled through population measures.

**Model specification**

Treatment: `Exposed_i` equals 1 for suburbs with early LXRP construction or completions by 2017, and 0 for LXRP-listed suburbs not yet exposed as of 2017.

Outcome: The natural logarithm of median house prices in a suburb-year.

Comparison group (Control): LXRP-listed suburbs not yet exposed by 2017.

Time period (2015 to 2024): Pre-treatment years are 2015-2017; post-treatment years are 2018 onward.

**Regression equation**:

$$\log(\text{price}_{it}) = \beta_0 + \beta_1 \text{Exposed}_i + \beta_2 \text{Post2017}_t + \delta(\text{Exposed}_i \times \text{Post2017}_t) + \gamma \log(\text{pop}_{it}) + \varepsilon_{it}$$

where:
- $\beta_1$ captures time-invariant level differences between early-exposed and not-yet-exposed suburbs prior to 2017.
- $\beta_2$ captures common post-2017 shifts in log prices affecting all suburbs.
- $\delta$ is the DiD estimator: the average causal effect of early LXRP exposure on log prices after 2017.
- $\gamma$ is the elasticity of price with respect to population (time-varying control).
- $\varepsilon_{it}$ is the idiosyncratic error term.

**Sample and regressors**

The sample includes LXRP-listed Melbourne suburbs observed over 2015-2024 with house-price and population data. Early-exposed suburbs are those with construction or completions by 2017; not-yet-exposed suburbs serve as controls.

**Error structure and interpretation**

Standard errors are clustered at the suburb level to account for serial correlation and repeated observations. Because the outcome is in logs, $e^{\delta}-1$ gives the proportional change in prices; for small $\delta$, $100\times\delta$ approximates the percent change.
