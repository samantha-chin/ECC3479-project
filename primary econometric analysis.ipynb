# Primary Econometric Analysis


**Analysis type**: Causal

**Identification strategy and assumptions**

We rely on the difference‑in‑differences identification strategy, which assumes that in the absence of LXRP, house‑price trends in treated suburbs would have evolved in parallel to those in untreated suburbs (parallel trends assumption).

We approximate this by:

- Defining treated suburbs as those within a fixed distance threshold of any completed or under‑construction LXRP site.
- Using multiple years of pre‑ and post‑treatment data.
- Including fixed‑effects‑style structure (e.g., year dummies or year‑period dummies) in the regression to soak up common time shocks.

**Model specification**

Treatment: A binary indicator equal to 1 if a suburb is within a pre-defined distance (≤ 2 km) of an LXRP site, and 0 otherwise. This captures which suburbs received the policy intervention.

Outcome: The natural logarithm of median house prices in a suburb in a given year. We use the log transformation to interpret the DiD coefficient as a percentage change in prices.

Comparison group (Control): Suburbs that are not within the distance threshold of any LXRP site. These suburbs serve as the counterfactual baseline to isolate the LXRP effect from general market trends.

Time period (2015 to 2024): The pre-treatment period (before LXRP projects commence) provides a baseline for parallel trends assessment and the post-treatment period captures the policy impact.

**Regression equation**:

$$\log(\text{price}_{it}) = \beta_0 + \beta_1 \text{Treat}_i + \beta_2 \text{Post}_t + \delta(\text{Treat}_i \times \text{Post}_t) + \gamma \log(\text{pop}_{it}) + \varepsilon_{it}$$

where:
- $\beta_1$ captures time‑invariant level differences between treated and control suburbs prior to the policy.
- $\beta_2$ captures common post‑treatment shifts in log prices affecting all suburbs (market‑wide shocks).
- $\delta$ is the difference‑in‑differences estimator: the average causal effect of LXRP on log prices for treated suburbs in the post period.
- $\gamma$ is the elasticity of price with respect to population (we include $\log(\text{pop}_{it})$ as a time‑varying control).
- $\varepsilon_{it}$ is the idiosyncratic error term.

**Sample and regressors**

Our sample consists of selected Melbourne suburbs observed over multiple years before and after the LXRP projects, with house‑price and population data available for each year. Treated suburbs are those within a pre‑defined distance (≤ 2 km) of the nearest LXRP site; all other suburbs in the sample are treated as controls.

**Error structure and justification**

We report coefficients with standard errors clustered at the suburb level to account for serial correlation and repeated observations.

Rationale and interpretation notes:
- The parameter $\delta$ identifies the average treatment effect on log prices for treated suburbs. Because the dependent variable is in logs, $e^{\delta}-1$ gives the proportional change in price; for small $\delta$, $100\times\delta$ approximates the percent change.
- We include $\log(\text{pop}_{it})$ as a time‑varying control to account for population dynamics that may mechanically affect median prices.
- If we instead include suburb fixed effects (unit dummies), the time‑invariant $\text{Treat}_i$ term would be collinear and thus absorbed; in that case the DiD estimator is obtained from the interaction term together with time dummies. Both formulations are common — we present the native interaction specification here for clarity and report alternative fixed‑effects specifications in robustness checks.
