Table 1 reports summary statistics for the final analysis sample. The average median house price is approximately $815,015, while the median is $720,000, showing that house prices are right-skewed. The sample also contains substantial variation in house prices, ranging from $177,000 to $2.65 million. This supports using log house prices in the regression, since the log transformation reduces the influence of very high-price suburbs. The average suburb population is 14,571, with a median of 13,738, showing that population size also varies across suburbs and should be included as a control.

```python
from pathlib import Path
import sys

import pandas as pd

project_root = Path.cwd()
sys.path.append(str(project_root / "code"))

from robustness_alternative_inference import build_panel


# Build analysis panel
df = build_panel()


# Create summary statistics table
summary_stats = pd.DataFrame(
    {
        "Variable": [
            "Median house price",
            "Log median house price",
            "Population",
            "Log population",
        ],
        "Mean": [
            df["price"].mean(),
            df["log_price"].mean(),
            df["population"].mean(),
            df["log_population"].mean(),
        ],
        "Median": [
            df["price"].median(),
            df["log_price"].median(),
            df["population"].median(),
            df["log_population"].median(),
        ],
        "Std. Dev.": [
            df["price"].std(),
            df["log_price"].std(),
            df["population"].std(),
            df["log_population"].std(),
        ],
        "Min": [
            df["price"].min(),
            df["log_price"].min(),
            df["population"].min(),
            df["log_population"].min(),
        ],
        "Max": [
            df["price"].max(),
            df["log_price"].max(),
            df["population"].max(),
            df["log_population"].max(),
        ],
    }
)


# Format values
def format_number(row, col):
    value = row[col]

    if row["Variable"] in ["Median house price", "Population"]:
        return f"{value:,.0f}"

    return f"{value:.3f}"


summary_stats_clean = summary_stats.copy()

for col in ["Mean", "Median", "Std. Dev.", "Min", "Max"]:
    summary_stats_clean[col] = summary_stats_clean.apply(
        lambda row: format_number(row, col),
        axis=1,
    )


# Display clean table
display(summary_stats_clean)


# Save table
results_dir = project_root / "results"
results_dir.mkdir(exist_ok=True)

summary_stats_clean.to_csv(
    results_dir / "summary_statistics.csv",
    index=False,
)

print("Saved to results/summary_statistics.csv")
```
