```python
import pandas as pd

raw = pd.read_csv("all_rows_G02.csv", header=None, dtype=str)

raw = raw.replace(r"^\s*$", pd.NA, regex=True)

left = raw[[0, 1]].rename(columns={0: "metric", 1: "value"})
right = raw[[3, 4]].rename(columns={3: "metric", 4: "value"})
tidy = pd.concat([left, right], ignore_index=True)

tidy["metric"] = tidy["metric"].astype("string").str.strip()
tidy["value"] = tidy["value"].astype("string").str.strip()

tidy = tidy.dropna(subset=["metric", "value"])
tidy = tidy[~tidy["metric"].str.contains("ABS|Census|List of tables|G02|excludes|applicable", case=False, na=False)]

tidy["value"] = pd.to_numeric(tidy["value"], errors="coerce")
tidy = tidy.dropna(subset=["value"]).drop_duplicates(subset=["metric"]).reset_index(drop=True)

print("NaN left:", int(tidy.isna().sum().sum()))
display(tidy)

tidy.to_csv("all_rows_G02_clean.csv", index=False)
```
